```julia
#tablice wielowymiarowe w Julii
Asmall=[[1.0 0.0]; [0.0 1.0]]
Bsmall=Asmall
#size(B,1)
```

```julia
# mnożenie macierzy - wersja naiwna
function naive_multiplication(A,B)
C=zeros(Float64,size(A,1),size(B,2))
  for i=1:size(A,1)
    for j=1:size(B,2)
        for k=1:size(A,2)
            C[i,j]=C[i,j]+A[i,k]*B[k,j]
        end
    end
end
C
end
```

```julia
#kompilacja
naive_multiplication(Asmall,Bsmall)
```

```julia
#kompilacja funkcji BLASowej do mnożenia macierzy
#https://docs.julialang.org/en/stable/stdlib/linalg/#BLAS-Functions-1
Asmall*Bsmall
```

```julia
# poprawiona funkcja korzytająca z powyższego oraz z faktu, że
#można zmieniać kolejność operacji dodawania (a co za tym idzie kolejnosc petli).
function better_multiplication( A,B )
C=zeros(Float64,size(A,1),size(B,2))
  for j=1:size(B,2)
    for k=1:size(A,2)
        for i=1:size(A,1)
            C[i,j]=C[i,j]+A[i,k]*B[k,j]
        end
    end
end
C
end
```

```julia
better_multiplication(Asmall, Bsmall)
```

```julia
x_iter = 100:100:1000
naive = []
better = []
blas = []
for i = x_iter
    A = rand(i, i)
    B = rand(i, i)
    append!(naive, @elapsed naive_multiplication(A, B))
    append!(better, @elapsed better_multiplication(A, B))
    append!(blas, @elapsed A * B)
end

scatter(x_iter, [naive, better, blas], labels=["naive" "better" "blas"], xlabel="Sample size", ylabel="Time [ms]")
```

# 3 - fitting curves

```julia
data_naive_fit = polyfit(data[:size], data[:naive], 3)
data_better_fit = polyfit(data[:size], data[:better], 3)
data_blas_fit = polyfit(data[:size], data[:blas], 3)
data_optimized_naive_fit = polyfit(data_optimized[:size], data_optimized[:naive], 3)
data_optimized_better_fit = polyfit(data_optimized[:size], data_optimized[:better], 3)
data_optimized_blas_fit = polyfit(data_optimized[:size], data_optimized[:blas], 3)
naive_fit = polyfit(x_iter, naive, 3)
better_fit = polyfit(x_iter, better, 3)
blas_fit = polyfit(x_iter, blas, 3)
```

```julia
scatter(data[:size], data[:naive], labels=["c naive"], xlabel="Sample size", ylabel="Time [ms]", color="red")
xs = 0:100:1000
ys = map(x -> polyval(data_naive_fit, x), xs);
plot!(xs, ys, linewidth=4, labels="c naive fit", linecolor="red")
```

```julia
function draw(polynomial, data, label)
    scatter(x_iter, data, labels=label, xlabel="Sample size", ylabel="Time [ms]", legend=:topleft)
    xs = 0:100:1000
    ys = map(x -> polyval(polynomial, x), xs);
    plot!(xs, ys, linewidth=4, labels=label)
end
```

```julia
function draw!(polynomial, data, label)
    scatter!(x_iter, data, labels=label, xlabel="Sample size", ylabel="Time [ms]")
    xs = 0:100:1000
    ys = map(x -> polyval(polynomial, x), xs);
    plot!(xs, ys, linewidth=4, labels=label)
end
```
