# Zadanie 4
Wymyślić własny przykład algorytmu niestablinego numerycznie. 
1. Zademostrować wersję niestabilną, pokazać, że działa źle.
2. Wyjaśnić, dlaczego działa źle.
3. Zademonstować wersję stabilną.
4. Język dowolny.

Prostym przykładem będzie "rekurencyjna" całka (obliczymy ją dla $n \in \{0, 1, ..., 6\}$): 
$$y_n=\int_{0}^{1} \frac{x^n}{x + 7} dx$$
Możemy się spodziewać, że całki na tym przedziale dla liczb dodatnich będą dodatnie, gdyż same funkcje nie przyjmują wartości dodatnich na tym przedziale.  Wyprowadźmy zatem wzór rekurencyjny:
$$y_n + 7y_{n-1}=\int_{0}^{1}\frac{x^n+7x^{n-1}}{x+7} dx = \int_{0}^{1}\frac{x^{n-1}(x + 7)}{x + 7}=\int_{0}^{1}x^{n-1}dx=\frac{1}{n}$$
Stąd:
$$y_n = \frac{1}{n} - 7y_{n-1}$$
Oraz:
$$y_{n-1}=\frac{1}{7n}-\frac{1}{7}y_n$$

Zacznijmy od obliczenia pierwszej wartości $y_n$ (dla n = 0) - algorytmem podstawowym (niestabilnym)

```julia
from_x = 0
to_x = 1
y_0 = log(to_x + 7) - log(from_x + 7)
println(y_0)
y = [1 - 7*y_0]
for n = 1:5
    push!(y, 1/n - 7*y[n])
end
y
```

Ze wzoru rekurencyjnego (wzór na wyraz $y_{n-1}$) widzimy, że błąd za każdym razem dzielony będzie przez 7.  
Widząc przyjęty przedział oraz wzór funkcji, widzimy, że funkcja powinna maleć wraz ze wzrostem n.

```julia
y_6 = 1/(7*7)*(1+1/7)
println(y_6)

y = [y_6]

for n=0:5
    prepend!(y, 1/(7*(6-n)) - (1/7)*y[1])
end
y
```

Widzimy zatem, że otrzymane wartości są malejące, zatem drugi algorytm jest stabilniejszy od pierwszego.
